```python
def euc(x, y):
    
    global x_test
    
    ans = 0
    
    for column in x_test.columns:
        ans += (x[column] - y[column]) ** 2
    
    return math.sqrt(ans)
```

```python
def returnDistAndind(test_index):
    
    '''For an index in the test file, this function will return the distances of the element at that index with each
    element in x_train file along with the corresponding outcome from y_train file'''
    
    global x_train
    global x_test
    global y_train
    
    curr_row_test = x_test.iloc[test_index]
    
    array = []
    
    for i in range(x_train.shape[0]):
        curr_row_train = x_train.iloc[i]
        
        array.append((euc(curr_row_test, curr_row_train), y_train.iloc[i]['Outcome']))
    return array
    
```

```python
def predictor(list_1):
    
    '''This function accepts the output returned by the above function and returns the count of 0s and 1s as matched by knn'''
    
    k = 11
    
    sorted_nmpy_arr = sorted(list_1, key=lambda x : x[0])[:k]
    res = [tt[1] for tt in sorted_nmpy_arr]
    
    return (np.unique(res, return_counts = 1))
    
```

```python
with open('y_test.csv', 'w') as file:
    file.write('Outcome\n')
    for j in range(x_test.shape[0]):
        # This loop runs for each element in the x_test file
        distances_with_outcomes = returnDistAndind(j)
        counts = predictor(distances_with_outcomes)
        max_index = counts[1].argmax()
        final_outcome = counts[0][max_index]
        file.write(str(final_outcome))
        file.write('\n')
```
